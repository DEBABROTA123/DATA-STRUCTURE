#1. String Slicing
String slicing is a technique used to extract a portion of a string based on specified indices. In Python, strings are indexed starting from 0, and negative indices can be used to access elements from the end.

```python
string = "Hello, World!"
print(string[0:5])       # Output: 'Hello' (substring from index 0 to 4)
print(string[7:12])      # Output: 'World' (substring from index 7 to 11)
print(string[::2])       # Output: 'Hlo ol!' (every second character)
print(string[::-1])      # Output: '!dlroW ,olleH' (reversed string)
```

**2. Key Features of Lists in Python**

Ordered: Elements in a list are stored in a specific order.

Mutable: You can change the elements of a list after it's created.

Dynamic: Lists can grow or shrink in size as needed.

Heterogeneous: Elements in a list can be of different data types.

Indexing and Slicing: You can access elements using indices and perform slicing operations.

```python
#access, modify, and delete elements in a list with examples.

# Accessing Elements in a Numeric List
numbers = [10, 20, 30, 40, 50]

# Accessing using positive and negative indices
print(numbers[0])    # Output: 10
print(numbers[-1])   # Output: 50

# Accessing using slicing
print(numbers[1:4])  # Output: [20, 30, 40]
print(numbers[::2])  # Output: [10, 30, 50]  (every second element)
# Modifying Elements in a Numeric List
numbers[2] = 35      # Change element at index 2
print(numbers)       # Output: [10, 20, 35, 40, 50]

# Modifying multiple elements using slicing
numbers[1:3] = [25, 45]  # Replace elements at index 1 and 2
print(numbers)           # Output: [10, 25, 45, 40, 50]
# Deleting Elements in a Numeric List
del numbers[3]             # Delete element at index 3
print(numbers)             # Output: [10, 25, 45, 50]

numbers.remove(45)         # Remove the value 45 from the list
print(numbers)             # Output: [10, 25, 50]

popped_number = numbers.pop()  # Remove and return the last element
print(popped_number)           # Output: 50
print(numbers)                 # Output: [10, 25]
```

**Tuples**

Immutable: Tuples, once created, cannot be changed. Think of them as the unshakeable guardians of data integrity.

Syntax: Defined using parentheses: (1, 2, 3).

Performance: Faster in terms of iteration.

Use Cases: Ideal for fixed collections of items, such as coordinates or records that shouldn't change.

**Lists**

Mutable: Lists are flexible and can be altered after creation. Add, remove, change—no limits!

Syntax: Defined using square brackets: [1, 2, 3].

Performance: Slower than tuples due to the ability to be modified.

Use Cases: Perfect for collections of items that need to be modified, like todo lists or dynamic datasets.

```python
# Tuples
tuple_example = (1, 2, 3)
# Trying to change a tuple value will throw an error
# tuple_example[0] = 0  # This will raise a TypeError

# Lists
list_example = [1, 2, 3]
list_example[0] = 0  # Lists can be modified freely
```

**5. Key Features of Sets and Examples**

Unique: A set cannot contain duplicate elements.

Mutable: You can add or remove elements from a set.

Heterogeneous: Elements can be of different data types.

```python
7. Describe how to add, modify, and delete items in a dictionary with examples.


# Creating a dictionary for a product
product_dict = {
    'name': 'Basmati Rice',
    'price': 120,  # Price in INR
    'quantity': 50, # Quantity in kg
    'brand': 'Dawar'
}

# Adding items
product_dict['origin'] = 'Punjab'           # Add origin of the product
print("After adding origin:", product_dict)

# Modifying items
product_dict['price'] = 130                  # Update price
print("After modifying price:", product_dict)

# Deleting items
del product_dict['quantity']                 # Delete quantity
print("After deleting quantity:", product_dict)

# Using pop() to delete and return a value
removed_brand = product_dict.pop('brand')    # Remove brand and return its value
print("Removed brand:", removed_brand)         # Output: Dawar
print("Dictionary after pop:", product_dict)   # Remaining items
```

```python
# Importance of Dictionary Keys Being Immutable

# In Python, dictionary keys must be immutable types, such as strings, numbers, or tuples.
# This is important because immutable keys ensure that the key-value mapping remains consistent.

# Example : Using a string as a key (immutable)
my_dict = {'name': 'Alice', 'age': 30}
print(my_dict['name'])  # Output: Alice



# Why immutability matters:
# 1. **Consistency**: If keys were mutable, changing a key after its insertion would break the dictionary's structure.
# 2. **Hashing**: Dictionaries use hash tables for efficient key lookups. Mutable objects cannot be hashed consistently.
# 3. **Performance**: Immutable keys provide faster access and better memory management.

# Conclusion: Using immutable keys ensures reliable and efficient key-value storage and retrieval in dictionaries.
```
